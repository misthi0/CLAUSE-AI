```python
# LANGGRAPH BASICS - STEP 2: Define Shared Graph State

print("\n" + "="*70)
print("LANGGRAPH BASICS - STEP 2: DEFINE SHARED GRAPH STATE")
print("="*70)

class GraphState(TypedDict):
    query: str
    contract_text: str
    legal: Dict[str, Any]
    compliance: Dict[str, Any]
    finance: Dict[str, Any]
    operations: Dict[str, Any]
    execution_order: List[str]
    timestamps: Dict[str, float]

print("\nGraphState defined with fields:")
print("   - query: str (user query)")
print("   - contract_text: str (input contract)")
print("   - legal: Dict (Legal agent output)")
print("   - compliance: Dict (Compliance agent output)")
print("   - finance: Dict (Finance agent output)")
print("   - operations: Dict (Operations agent output)")
print("   - execution_order: List (track execution)")
print("   - timestamps: Dict (performance metrics)")

print("\n" + "="*70)
print("STEP 2 COMPLETE")
print("="*70)
```

```python
# LANGGRAPH BASICS - STEP 3: Define Agent Nodes

print("\n" + "="*70)
print("LANGGRAPH BASICS - STEP 3: DEFINE AGENT NODES")
print("="*70)

def legal_node(state: GraphState) -> GraphState:
    print("\n" + "="*70)
    print("LEGAL AGENT EXECUTING...")
    print("="*70)
    print(f"Query: {state['query']}")
    print(f"Execution order so far: {state['execution_order']}")
    
    start_time = time.time()
    
    # Use pre-loaded legal output
    state["legal"] = legal_output
    state["execution_order"].append("Legal")
    state["timestamps"]["legal"] = time.time() - start_time
    
    print(f"Legal Agent Complete ({state['timestamps']['legal']:.2f}s)")
    print(f"   Clauses extracted: {len(state['legal']['extracted_clauses'])}")
    print(f"   Risk level: {state['legal']['risk_level']}")
    print(f"Updated order: {state['execution_order']}")
    print("="*70)
    return state

def compliance_node(state: GraphState) -> GraphState:
    print("\n" + "="*70)
    print("COMPLIANCE AGENT EXECUTING...")
    print("="*70)
    print(f"Query: {state['query']}")
    print(f"Execution order so far: {state['execution_order']}")
    
    start_time = time.time()
    
    # Use pre-loaded compliance output
    state["compliance"] = compliance_output
    state["execution_order"].append("Compliance")
    state["timestamps"]["compliance"] = time.time() - start_time
    
    print(f"Compliance Agent Complete ({state['timestamps']['compliance']:.2f}s)")
    print(f"   Clauses extracted: {len(state['compliance']['extracted_clauses'])}")
    print(f"   Risk level: {state['compliance']['risk_level']}")
    print(f"Updated order: {state['execution_order']}")
    print("="*70)
    return state

def finance_node(state: GraphState) -> GraphState:
    print("\n" + "="*70)
    print("FINANCE AGENT EXECUTING...")
    print("="*70)
    print(f"Query: {state['query']}")
    print(f"Execution order so far: {state['execution_order']}")
    
    start_time = time.time()
    
    # Use pre-loaded finance output
    state["finance"] = finance_output
    state["execution_order"].append("Finance")
    state["timestamps"]["finance"] = time.time() - start_time
    
    print(f"Finance Agent Complete ({state['timestamps']['finance']:.2f}s)")
    print(f"   Clauses extracted: {len(state['finance']['extracted_clauses'])}")
    print(f"   Risk level: {state['finance']['risk_level']}")
    print(f"Updated order: {state['execution_order']}")
    print("="*70)
    return state

def operations_node(state: GraphState) -> GraphState:
    print("\n" + "="*70)
    print("OPERATIONS AGENT EXECUTING...")
    print("="*70)
    print(f"Query: {state['query']}")
    print(f"Execution order so far: {state['execution_order']}")
    
    start_time = time.time()
    
    # Use pre-loaded operations output
    state["operations"] = operations_output
    state["execution_order"].append("Operations")
    state["timestamps"]["operations"] = time.time() - start_time
    
    print(f"Operations Agent Complete ({state['timestamps']['operations']:.2f}s)")
    print(f"   Clauses extracted: {len(state['operations']['extracted_clauses'])}")
    print(f"   Risk level: {state['operations']['risk_level']}")
    print(f"Updated order: {state['execution_order']}")
    print("="*70)
    return state

print("\nAll 4 agent nodes defined with logging:")
print("   - legal_node()")
print("   - compliance_node()")
print("   - finance_node()")
print("   - operations_node()")

print("\n" + "="*70)
print("STEP 3 COMPLETE")
print("="*70)
```

```python
# LANGGRAPH BASICS - STEP 4: Build Graph Skeleton

print("\n" + "="*70)
print("LANGGRAPH BASICS - STEP 4: BUILD GRAPH SKELETON")
print("="*70)

workflow = StateGraph(GraphState)

print("\nStateGraph initialized")
print("   Type: StateGraph")
print("   State Schema: GraphState")

print("\n" + "="*70)
print("STEP 4 COMPLETE")
print("="*70)
```

```python
# LANGGRAPH BASICS - STEP 5: Define Edges (Simple Flow)

print("\n" + "="*70)
print("LANGGRAPH BASICS - STEP 5: DEFINE EDGES (SIMPLE FLOW)")
print("="*70)

# Add nodes to graph
workflow.add_node("legal_agent", legal_node)
workflow.add_node("compliance_agent", compliance_node)
workflow.add_node("finance_agent", finance_node)
workflow.add_node("operations_agent", operations_node)

print("\nNodes added to graph:")
print("   - legal_agent")
print("   - compliance_agent")
print("   - finance_agent")
print("   - operations_agent")

# Set entry point and edges (Legal first)
workflow.set_entry_point("legal_agent")
workflow.add_edge("legal_agent", "compliance_agent")
workflow.add_edge("compliance_agent", "finance_agent")
workflow.add_edge("finance_agent", "operations_agent")
workflow.add_edge("operations_agent", END)

print("\nExecution flow defined:")
print("\n   START")
print("     |")
print("   Legal Agent")
print("     |")
print("   Compliance Agent")
print("     |")
print("   Finance Agent")
print("     |")
print("   Operations Agent")
print("     |")
print("   END")

print("\n" + "="*70)
print("STEP 5 COMPLETE")
print("="*70)
```

```python
# LANGGRAPH BASICS - STEP 6: Compile Graph

print("\n" + "="*70)
print("LANGGRAPH BASICS - STEP 6: COMPILE GRAPH")
print("="*70)

app = workflow.compile()

print("\nGraph compiled successfully!")
print("   Status: Ready for execution")
print("   Agents: 4 (Legal, Compliance, Finance, Operations)")
print("   Flow: Sequential")

print("\n" + "="*70)
print("STEP 6 COMPLETE")
print("="*70)
```

```python
# MULTI-AGENT GRAPH - STEP 4: Add Nodes to Graph

print("\n" + "="*70)
print("MULTI-AGENT GRAPH - STEP 4: ADD NODES TO GRAPH")
print("="*70)

graph.add_node("legal_agent", legal_node)
graph.add_node("compliance_agent", compliance_node)
graph.add_node("finance_agent", finance_node)
graph.add_node("operations_agent", operations_node)

print("\nAll nodes added to graph:")
print("   - legal_agent -> legal_node()")
print("   - compliance_agent -> compliance_node()")
print("   - finance_agent -> finance_node()")
print("   - operations_agent -> operations_node()")

print("\n" + "="*70)
print("STEP 4 COMPLETE")
print("="*70)
```

```python
# CONDITIONAL ROUTING - STEP 1: Define Routing Function

print("="*70)
print("CONDITIONAL ROUTING - STEP 1: DEFINE ROUTING FUNCTION")
print("="*70)

def route_to_agent(state: MultiAgentGraphState) -> str:
    """
    Route query to appropriate agent based on keywords.
    Returns the name of the agent node to execute.
    
    Args:
        state: Current graph state
        
    Returns:
        str: Agent node name or END
    """
    query = state["query"].lower()
    
    print(f"\nRouting query: '{state['query']}'")
    print("-" * 70)
    
    # Define routing keywords
    legal_keywords = ["termination", "breach", "governing law", "jurisdiction", 
                      "liability", "indemnification", "indemnity", "contract"]
    
    compliance_keywords = ["gdpr", "compliance", "data protection", "privacy", 
                          "audit", "regulatory", "soc2", "iso", "hipaa"]
    
    finance_keywords = ["payment", "penalty", "late fee", "invoice", "refund", 
                       "pricing", "cost", "financial"]
    
    operations_keywords = ["sla", "timeline", "milestone", "deliverable", 
                          "uptime", "performance", "delivery"]
    
    # Check which keywords match
    legal_match = any(kw in query for kw in legal_keywords)
    compliance_match = any(kw in query for kw in compliance_keywords)
    finance_match = any(kw in query for kw in finance_keywords)
    operations_match = any(kw in query for kw in operations_keywords)
    
    # Route to first matching agent
    if legal_match:
        print("MATCHED: Legal Agent")
        print(f"   Keywords found: {[kw for kw in legal_keywords if kw in query]}")
        return "legal_agent"
    elif compliance_match:
        print("MATCHED: Compliance Agent")
        print(f"   Keywords found: {[kw for kw in compliance_keywords if kw in query]}")
        return "compliance_agent"
    elif finance_match:
        print("MATCHED: Finance Agent")
        print(f"   Keywords found: {[kw for kw in finance_keywords if kw in query]}")
        return "finance_agent"
    elif operations_match:
        print("MATCHED: Operations Agent")
        print(f"   Keywords found: {[kw for kw in operations_keywords if kw in query]}")
        return "operations_agent"
    else:
        print("NO MATCH: Ending execution")
        return END

print("\nroute_to_agent() function defined")
print("   - Matches keywords from query")
print("   - Returns agent node name")
print("   - Returns END if no match")

print("\n" + "="*70)
print("STEP 1 COMPLETE")
print("="*70)
```

```python
# CONDITIONAL ROUTING - STEP 2: Rebuild Graph with Conditional Entry

print("\n" + "="*70)
print("CONDITIONAL ROUTING - STEP 2: REBUILD GRAPH")
print("="*70)

from langgraph.graph import StateGraph, END, START

# Create new graph for conditional routing
conditional_graph = StateGraph(MultiAgentGraphState)

# Add all agent nodes
conditional_graph.add_node("legal_agent", legal_node)
conditional_graph.add_node("compliance_agent", compliance_node)
conditional_graph.add_node("finance_agent", finance_node)
conditional_graph.add_node("operations_agent", operations_node)

print("\nConditional graph created")
print("   All 4 agent nodes added")

print("\n" + "="*70)
print("STEP 2 COMPLETE")
print("="*70)
```

```python
# CONDITIONAL ROUTING - STEP 3: Conditional Entry Point

print("\n" + "="*70)
print("CONDITIONAL ROUTING - STEP 3: CONDITIONAL ENTRY POINT")
print("="*70)

# Add conditional edges from START
conditional_graph.add_conditional_edges(
    START,
    route_to_agent,
    {
        "legal_agent": "legal_agent",
        "compliance_agent": "compliance_agent",
        "finance_agent": "finance_agent",
        "operations_agent": "operations_agent",
        END: END
    }
)

print("\nConditional entry point configured")
print("   START -> route_to_agent() -> Agent")
print("\nPossible routes:")
print("   - Legal Agent")
print("   - Compliance Agent")
print("   - Finance Agent")
print("   - Operations Agent")
print("   - END (no match)")

print("\n" + "="*70)
print("STEP 3 COMPLETE")
print("="*70)
```

```python
# CONDITIONAL ROUTING - STEP 4 & 5: Agent -> END Edges

print("\n" + "="*70)
print("CONDITIONAL ROUTING - STEP 4 & 5: AGENT -> END EDGES")
print("="*70)

# Each agent goes directly to END (no sequential flow)
conditional_graph.add_edge("legal_agent", END)
conditional_graph.add_edge("compliance_agent", END)
conditional_graph.add_edge("finance_agent", END)
conditional_graph.add_edge("operations_agent", END)

print("\nAgent -> END edges configured")
print("   Legal Agent -> END")
print("   Compliance Agent -> END")
print("   Finance Agent -> END")
print("   Operations Agent -> END")

print("\nGraph flow:")
print("\n   START")
print("     |")
print("   Route Function")
print("     |")
print("   +--+--+--+")
print("   |  |  |  |")
print("   L  C  F  O  (Only ONE executes)")
print("   |  |  |  |")
print("   +--+--+--+")
print("     |")
print("   END")

# Compile conditional graph
app_conditional = conditional_graph.compile()

print("\nConditional graph compiled successfully!")

print("\n" + "="*70)
print("STEP 4 & 5 COMPLETE")
print("="*70)
```

Agent-to-Agent Communication & Validation Logic

```python
# AGENT COMMUNICATION - STEP 1: Updated Graph State

print("="*70)
print("AGENT COMMUNICATION - STEP 1: UPDATED GRAPH STATE")
print("="*70)

from typing import TypedDict, Dict, Any, List

class CollaborativeGraphState(TypedDict):
    query: str
    contract_text: str
    memory: List[Dict[str, Any]]  # Shared memory across agents
    validation_notes: List[str]   # NEW: Validation messages
    legal: Dict[str, Any]
    compliance: Dict[str, Any]
    finance: Dict[str, Any]
    operations: Dict[str, Any]
    execution_order: List[str]
    timestamps: Dict[str, float]

print("\nCollaborativeGraphState defined:")
print("   - query: str")
print("   - contract_text: str")
print("   - memory: List[Dict] (shared across agents)")
print("   - validation_notes: List[str] <- NEW: Inter-agent validation")
print("   - legal: Dict")
print("   - compliance: Dict")
print("   - finance: Dict")
print("   - operations: Dict")
print("   - execution_order: List")
print("   - timestamps: Dict")

print("\nValidation Notes Purpose:")
print("   - Agents can flag conflicts")
print("   - Agents can validate each other's findings")
print("   - Enables collaborative analysis")

print("\n" + "="*70)
print("STEP 1 COMPLETE")
print("="*70)
```

```python
# AGENT COMMUNICATION - STEP 2: Initialize State

print("\n" + "="*70)
print("AGENT COMMUNICATION - STEP 2: INITIALIZE STATE")
print("="*70)

initial_collaborative_state = {
    "query": "Review compliance, finance, and legal issues",
    "contract_text": "Full contract text...",
    "memory": [],
    "validation_notes": [],  # NEW: Empty validation notes
    "legal": {},
    "compliance": {},
    "finance": {},
    "operations": {},
    "execution_order": [],
    "timestamps": {}
}

print("\nInitial collaborative state:")
print(f"   query: {initial_collaborative_state['query']}")
print(f"   memory: {initial_collaborative_state['memory']}")
print(f"   validation_notes: {initial_collaborative_state['validation_notes']}")

print("\n" + "="*70)
print("STEP 2 COMPLETE")
print("="*70)
```

```python
# AGENT COMMUNICATION - STEP 3: Compliance Agent

print("\n" + "="*70)
print("AGENT COMMUNICATION - STEP 3: COMPLIANCE AGENT (WRITES)")
print("="*70)

import time
from datetime import datetime

def compliance_node_collaborative(state: CollaborativeGraphState) -> CollaborativeGraphState:
    print("\n" + "="*70)
    print("COMPLIANCE AGENT EXECUTING (COLLABORATIVE MODE)...")
    print("="*70)
    print(f"Query: {state['query']}")
    print(f"Current memory entries: {len(state['memory'])}")
    
    start_time = time.time()
    
    # Execute compliance analysis
    output = compliance_output
    state["compliance"] = output
    state["execution_order"].append("Compliance")
    state["timestamps"]["compliance"] = time.time() - start_time
    
    # Write to shared memory
    memory_entry = {
        "agent": "compliance",
        "timestamp": datetime.now().isoformat(),
        "findings": output["extracted_clauses"],
        "risk_level": output["risk_level"],
        "confidence": output["confidence"]
    }
    state["memory"].append(memory_entry)
    
    print(f"\nCompliance Agent Complete ({state['timestamps']['compliance']:.4f}s)")
    print(f"   Findings: {len(output['extracted_clauses'])} clauses")
    print(f"   Risk: {output['risk_level'].upper()}")
    print(f"   Written to memory: {len(output['extracted_clauses'])} findings")
    
    print(f"\nMemory after Compliance:")
    for i, entry in enumerate(state['memory'], 1):
        print(f"   {i}. {entry['agent']}: {len(entry.get('findings', []))} findings")
    
    print("="*70)
    return state

print("\nCompliance node defined:")
print("   - Executes compliance analysis")
print("   - Writes findings to shared memory")
print("   - Makes data available for other agents")

print("\n" + "="*70)
print("STEP 3 COMPLETE")
print("="*70)
```

```python
# AGENT COMMUNICATION - STEP 4: Finance Agent (Reads Compliance)

print("\n" + "="*70)
print("AGENT COMMUNICATION - STEP 4: FINANCE AGENT (READS & VALIDATES)")
print("="*70)

def finance_node_collaborative(state: CollaborativeGraphState) -> CollaborativeGraphState:
    print("\n" + "="*70)
    print("FINANCE AGENT EXECUTING (COLLABORATIVE MODE)...")
    print("="*70)
    print(f"Query: {state['query']}")
    print(f"Current memory entries: {len(state['memory'])}")
    
    # Read compliance findings from memory
    compliance_findings = [m for m in state["memory"] if m["agent"] == "compliance"]
    
    if compliance_findings:
        print("\nReading Compliance Memory:")
        for finding in compliance_findings:
            print(f"   Agent: {finding['agent']}")
            print(f"   Findings: {len(finding.get('findings', []))} clauses")
            print(f"   Risk: {finding.get('risk_level', 'unknown').upper()}")
    
    start_time = time.time()
    
    # Execute finance analysis
    output = finance_output
    state["finance"] = output
    state["execution_order"].append("Finance")
    state["timestamps"]["finance"] = time.time() - start_time
    
    # Cross-validate with compliance findings
    if compliance_findings:
        validation_note = (
            f"Finance reviewed compliance findings for penalty conflicts. "
            f"Compliance identified {len(compliance_findings[0].get('findings', []))} issues, "
            f"Finance found {len(output['extracted_clauses'])} payment clauses."
        )
        state["validation_notes"].append(validation_note)
        print(f"\n   VALIDATION NOTE ADDED:")
        print(f"   {validation_note}")
    
    # Write to shared memory
    memory_entry = {
        "agent": "finance",
        "timestamp": datetime.now().isoformat(),
        "findings": output["extracted_clauses"],
        "risk_level": output["risk_level"],
        "confidence": output["confidence"],
        "validated_against": ["compliance"] if compliance_findings else []
    }
    state["memory"].append(memory_entry)
    
    print(f"\nFinance Agent Complete ({state['timestamps']['finance']:.4f}s)")
    print(f"   Findings: {len(output['extracted_clauses'])} clauses")
    print(f"   Risk: {output['risk_level'].upper()}")
    print(f"   Validated against: {memory_entry['validated_against']}")
    
    print(f"\nMemory after Finance:")
    for i, entry in enumerate(state['memory'], 1):
        print(f"   {i}. {entry['agent']}: {len(entry.get('findings', []))} findings")
    
    print("="*70)
    return state

print("\nFinance node defined:")
print("   - Reads compliance memory")
print("   - Validates findings")
print("   - Adds validation notes")
print("   - Writes own findings to memory")

print("\n" + "="*70)
print("STEP 4 COMPLETE")
print("="*70)
```

```python
# AGENT COMMUNICATION - STEP 5: Legal Agent (Final Validation)

print("\n" + "="*70)
print("AGENT COMMUNICATION - STEP 5: LEGAL AGENT (FINAL VALIDATION)")
print("="*70)

def legal_node_collaborative(state: CollaborativeGraphState) -> CollaborativeGraphState:
    print("\n" + "="*70)
    print("LEGAL AGENT EXECUTING (FINAL VALIDATION)...")
    print("="*70)
    print(f"Query: {state['query']}")
    print(f"Current memory entries: {len(state['memory'])}")
    print(f"Current validation notes: {len(state['validation_notes'])}")
    
    # Read all previous findings
    print("\nReading All Previous Findings:")
    compliance_findings = [m for m in state["memory"] if m["agent"] == "compliance"]
    finance_findings = [m for m in state["memory"] if m["agent"] == "finance"]
    
    if compliance_findings:
        print(f"   Compliance: {len(compliance_findings[0].get('findings', []))} clauses")
    if finance_findings:
        print(f"   Finance: {len(finance_findings[0].get('findings', []))} clauses")
    
    start_time = time.time()
    
    # Execute legal analysis
    output = legal_output
    state["legal"] = output
    state["execution_order"].append("Legal")
    state["timestamps"]["legal"] = time.time() - start_time
    
    # Final validation across all agents
    total_clauses = sum(len(m.get('findings', [])) for m in state["memory"])
    
    validation_note = (
        f"Legal performed final validation across all agents. "
        f"Total clauses identified: {total_clauses + len(output['extracted_clauses'])}. "
        f"Legal found {len(output['extracted_clauses'])} termination/liability clauses."
    )
    state["validation_notes"].append(validation_note)
    
    # Check for conflicts
    high_risk_agents = [m['agent'] for m in state["memory"] if m.get('risk_level') == 'high']
    if high_risk_agents:
        conflict_note = f"Legal identified high-risk findings from: {', '.join(high_risk_agents)}. Recommend legal review."
        state["validation_notes"].append(conflict_note)
        print(f"\n   CONFLICT DETECTED:")
        print(f"   {conflict_note}")
    
    print(f"\n   FINAL VALIDATION NOTE ADDED:")
    print(f"   {validation_note}")
    
    # Write to shared memory
    memory_entry = {
        "agent": "legal",
        "timestamp": datetime.now().isoformat(),
        "findings": output["extracted_clauses"],
        "risk_level": output["risk_level"],
        "confidence": output["confidence"],
        "validated_against": ["compliance", "finance"],
        "final_validation": True
    }
    state["memory"].append(memory_entry)
    
    print(f"\nLegal Agent Complete ({state['timestamps']['legal']:.4f}s)")
    print(f"   Findings: {len(output['extracted_clauses'])} clauses")
    print(f"   Risk: {output['risk_level'].upper()}")
    print(f"   Final validation: Complete")
    
    print(f"\nFinal Memory:")
    for i, entry in enumerate(state['memory'], 1):
        print(f"   {i}. {entry['agent']}: {len(entry.get('findings', []))} findings")
    
    print(f"\nAll Validation Notes:")
    for i, note in enumerate(state['validation_notes'], 1):
        print(f"   {i}. {note}")
    
    print("="*70)
    return state

print("\nLegal node defined:")
print("   - Reads all previous memory")
print("   - Performs final validation")
print("   - Identifies conflicts")
print("   - Adds comprehensive validation notes")

print("\n" + "="*70)
print("STEP 5 COMPLETE")
print("="*70)
```

```python
# AGENT COMMUNICATION - STEP 6: Build Collaborative Graph

print("\n" + "="*70)
print("AGENT COMMUNICATION - STEP 6: BUILD COLLABORATIVE GRAPH")
print("="*70)

from langgraph.graph import StateGraph, END

# Create collaborative graph
collab_graph = StateGraph(CollaborativeGraphState)

# Add nodes
collab_graph.add_node("compliance_agent", compliance_node_collaborative)
collab_graph.add_node("finance_agent", finance_node_collaborative)
collab_graph.add_node("legal_agent", legal_node_collaborative)

print("\nCollaborative graph created with 3 agents:")
print("   - compliance_agent (writes)")
print("   - finance_agent (reads compliance, writes)")
print("   - legal_agent (reads all, final validation)")

print("\n" + "="*70)
print("STEP 6 COMPLETE")
print("="*70)
```

```python
# AGENT COMMUNICATION - STEP 7: Compile Graph

print("\n" + "="*70)
print("AGENT COMMUNICATION - STEP 7: DEFINE EDGES & COMPILE")
print("="*70)

# Set execution flow
collab_graph.set_entry_point("compliance_agent")
collab_graph.add_edge("compliance_agent", "finance_agent")
collab_graph.add_edge("finance_agent", "legal_agent")
collab_graph.add_edge("legal_agent", END)

print("\nCollaborative execution flow:")
print("\n   START")
print("     |")
print("   Compliance (writes memory)")
print("     |")
print("   Finance (reads compliance, validates)")
print("     |")
print("   Legal (reads all, final validation)")
print("     |")
print("   END")

# Compile
app_collaborative = collab_graph.compile()

print("\nCollaborative graph compiled!")

print("\n" + "="*70)
print("STEP 7 COMPLETE")
print("="*70)
```

```python
# STEP 1: Quick Setup (No Model Loading - INSTANT)

print("="*70)
print("FAST SETUP: Compliance Pipeline")
print("="*70)

# Use keyword-based retrieval (no model needed)
def retrieve_compliance_chunks(text, keywords):
    """Fast keyword-based retrieval."""
    paragraphs = [p.strip() for p in text.split('\n\n') if p.strip()]
    
    retrieved_chunks = []
    for i, para in enumerate(paragraphs):
        relevance_score = sum(1 for kw in keywords if kw.lower() in para.lower())
        
        if relevance_score > 0:
            retrieved_chunks.append({
                "chunk_id": i,
                "text": para,
                "relevance_score": relevance_score,
                "matched_keywords": [kw for kw in keywords if kw.lower() in para.lower()]
            })
    
    retrieved_chunks.sort(key=lambda x: x['relevance_score'], reverse=True)
    return retrieved_chunks

print("\n✓ Fast retrieval function ready")
print("  Execution: INSTANT (no model loading)")

print("\n" + "="*70)
print("SETUP COMPLETE")
print("="*70)
```

```python
# STEP 6: Validate Compliance Output (FIXED)

print("\n" + "="*70)
print("COMPLIANCE PIPELINE - STEP 5: VALIDATE OUTPUT")
print("="*70)

# Define validation function
def validate_agent_output(agent_output, clause_type):
    """
    Validate agent output to match standard schema.
    """
    print(f"\nValidating output for clause_type: '{clause_type}'")
    
    try:
        # Validate required fields
        validated_output = {
            "clause_type": clause_type,
            "extracted_clauses": agent_output.get("extracted_clauses", []),
            "risk_level": agent_output.get("risk_level", "unknown").lower(),
            "confidence": float(agent_output.get("confidence", 0.0)),
            "evidence": agent_output.get("evidence", [])
        }
        
        # Validate risk_level values
        valid_risk_levels = ["low", "medium", "high", "unknown"]
        if validated_output["risk_level"] not in valid_risk_levels:
            print(f"   Warning: Invalid risk_level '{validated_output['risk_level']}', setting to 'unknown'")
            validated_output["risk_level"] = "unknown"
        
        # Validate confidence range
        if not (0.0 <= validated_output["confidence"] <= 1.0):
            print(f"   Warning: Confidence {validated_output['confidence']} out of range, clamping to [0.0, 1.0]")
            validated_output["confidence"] = max(0.0, min(1.0, validated_output["confidence"]))
        
        # Validate lists
        if not isinstance(validated_output["extracted_clauses"], list):
            validated_output["extracted_clauses"] = []
        if not isinstance(validated_output["evidence"], list):
            validated_output["evidence"] = []
        
        print("   Validation: SUCCESS")
        print(f"   Extracted clauses: {len(validated_output['extracted_clauses'])}")
        print(f"   Risk level: {validated_output['risk_level']}")
        print(f"   Confidence: {validated_output['confidence']:.2f}")
        
        return validated_output
        
    except Exception as e:
        print(f"   ERROR: Validation failed - {e}")
        return {
            "clause_type": clause_type,
            "extracted_clauses": [],
            "risk_level": "unknown",
            "confidence": 0.0,
            "evidence": [],
            "error": str(e)
        }

# Validate the compliance output
validated_compliance = validate_agent_output(
    compliance_pipeline_result,
    clause_type="Compliance"
)

print("\nValidation Complete")
print(f"   Status: SUCCESS")

print("\nValidated Output:")
print(f"   Clause Type: {validated_compliance['clause_type']}")
print(f"   Extracted Clauses: {len(validated_compliance['extracted_clauses'])}")
print(f"   Risk Level: {validated_compliance['risk_level'].upper()}")
print(f"   Confidence: {validated_compliance['confidence']:.0%}")
print(f"   Evidence Count: {len(validated_compliance['evidence'])}")

# Validation checks
validation_checks = {
    "has_clauses": len(validated_compliance['extracted_clauses']) > 0,
    "valid_risk_level": validated_compliance['risk_level'] in ['low', 'medium', 'high', 'unknown'],
    "valid_confidence": 0.0 <= validated_compliance['confidence'] <= 1.0,
    "has_evidence": len(validated_compliance['evidence']) > 0
}

print("\nValidation Checks:")
for check, passed in validation_checks.items():
    status = "PASS" if passed else "FAIL"
    print(f"   {check}: {status}")

all_passed = all(validation_checks.values())
print(f"\nOverall Validation: {'PASS' if all_passed else 'FAIL'}")

print("\n" + "="*70)
print("STEP 5 COMPLETE")
print("="*70)
```

```python
# STEP 7: Compliance Risk Summary

print("\n" + "="*70)
print("COMPLIANCE PIPELINE - STEP 6: RISK SUMMARY")
print("="*70)

def generate_compliance_risk_summary(validated_output):
    """Generate comprehensive compliance risk summary."""
    summary = {
        "overall_risk": validated_output['risk_level'],
        "confidence_score": validated_output['confidence'],
        "total_clauses_identified": len(validated_output['extracted_clauses']),
        "compliance_areas": [],
        "risk_factors": [],
        "recommendations": []
    }
    
    # Identify compliance areas
    compliance_areas = set()
    for clause in validated_output['extracted_clauses']:
        clause_lower = clause.lower()
        if 'gdpr' in clause_lower:
            compliance_areas.add("GDPR")
        if 'data protection' in clause_lower or 'privacy' in clause_lower:
            compliance_areas.add("Data Protection")
        if 'audit' in clause_lower:
            compliance_areas.add("Audit & Reporting")
        if 'security' in clause_lower:
            compliance_areas.add("Security Standards")
        if 'breach' in clause_lower:
            compliance_areas.add("Breach Notification")
        if 'certification' in clause_lower or 'soc2' in clause_lower or 'iso' in clause_lower:
            compliance_areas.add("Certifications")
    
    summary['compliance_areas'] = list(compliance_areas)
    
    # Risk factors and recommendations
    if validated_output['risk_level'] == 'high':
        summary['risk_factors'].append("High compliance risk identified")
        summary['recommendations'].append("Immediate legal review required")
        summary['recommendations'].append("Ensure GDPR compliance measures are in place")
    elif validated_output['risk_level'] == 'medium':
        summary['risk_factors'].append("Moderate compliance requirements")
        summary['recommendations'].append("Regular compliance audits recommended")
    else:
        summary['risk_factors'].append("Standard compliance requirements")
        summary['recommendations'].append("Maintain current compliance practices")
    
    if validated_output['confidence'] < 0.7:
        summary['risk_factors'].append(f"Low confidence score ({validated_output['confidence']:.0%})")
        summary['recommendations'].append("Manual review of compliance clauses recommended")
    
    return summary

# Generate risk summary
compliance_risk_summary = generate_compliance_risk_summary(validated_compliance)

print("\nCOMPLIANCE RISK SUMMARY")
print("="*70)

print(f"\nOverall Risk: {compliance_risk_summary['overall_risk'].upper()}")
print(f"Confidence Score: {compliance_risk_summary['confidence_score']:.0%}")
print(f"Total Clauses: {compliance_risk_summary['total_clauses_identified']}")

print(f"\nCompliance Areas Covered:")
if compliance_risk_summary['compliance_areas']:
    for area in compliance_risk_summary['compliance_areas']:
        print(f"   - {area}")
else:
    print("   - No specific areas identified")

print(f"\nRisk Factors:")
for factor in compliance_risk_summary['risk_factors']:
    print(f"   - {factor}")

print(f"\nRecommendations:")
for rec in compliance_risk_summary['recommendations']:
    print(f"   - {rec}")

print("\n" + "="*70)
print("STEP 6 COMPLETE")
print("="*70)
```

```python
# FINANCE PIPELINE - STEP 6: Finance Risk Summary

print("\n" + "="*70)
print("FINANCE PIPELINE - STEP 6: FINANCE RISK SUMMARY")
print("="*70)

def generate_finance_risk_summary(validated_output):
    """Generate comprehensive finance risk summary."""
    summary = {
        "overall_risk": validated_output['risk_level'],
        "confidence_score": validated_output['confidence'],
        "total_clauses_identified": len(validated_output['extracted_clauses']),
        "finance_areas": [],
        "risk_factors": [],
        "recommendations": [],
        "financial_impact": "Unknown"
    }
    
    # Identify finance areas
    finance_areas = set()
    penalty_count = 0
    default_count = 0
    
    for clause in validated_output['extracted_clauses']:
        clause_lower = clause.lower()
        
        if 'payment' in clause_lower or 'invoice' in clause_lower:
            finance_areas.add("Payment Terms")
        if 'penalty' in clause_lower or 'late fee' in clause_lower:
            finance_areas.add("Penalties")
            penalty_count += 1
        if 'refund' in clause_lower:
            finance_areas.add("Refund Policy")
        if 'default' in clause_lower:
            finance_areas.add("Payment Default")
            default_count += 1
        if 'reimbursement' in clause_lower or 'expense' in clause_lower:
            finance_areas.add("Expense Reimbursement")
        if 'billing' in clause_lower:
            finance_areas.add("Billing & Invoicing")
        if 'interest' in clause_lower:
            finance_areas.add("Interest Charges")
    
    summary['finance_areas'] = list(finance_areas)
    
    # Risk factors and financial impact
    if validated_output['risk_level'] == 'high':
        summary['risk_factors'].append(f"High financial risk identified ({penalty_count} penalty clauses)")
        summary['recommendations'].append("Review all penalty clauses carefully")
        summary['recommendations'].append("Negotiate payment terms if possible")
        summary['financial_impact'] = "HIGH - Multiple penalties and strict payment terms"
        
        if default_count > 0:
            summary['risk_factors'].append(f"Payment default consequences identified ({default_count} clauses)")
            summary['recommendations'].append("Ensure payment processes are robust")
    
    elif validated_output['risk_level'] == 'medium':
        summary['risk_factors'].append("Moderate financial obligations")
        summary['recommendations'].append("Monitor payment deadlines closely")
        summary['recommendations'].append("Set up automated payment reminders")
        summary['financial_impact'] = "MEDIUM - Standard payment terms with some penalties"
    
    else:
        summary['risk_factors'].append("Standard financial terms")
        summary['recommendations'].append("Maintain regular payment schedule")
        summary['financial_impact'] = "LOW - Favorable payment conditions"
    
    if validated_output['confidence'] < 0.7:
        summary['risk_factors'].append(f"Low confidence score ({validated_output['confidence']:.0%})")
        summary['recommendations'].append("Manual review of financial clauses recommended")
    
    return summary

# Generate finance risk summary
finance_risk_summary = generate_finance_risk_summary(validated_finance)

print("\nFINANCE RISK SUMMARY")
print("="*70)

print(f"\nOverall Risk: {finance_risk_summary['overall_risk'].upper()}")
print(f"Confidence Score: {finance_risk_summary['confidence_score']:.0%}")
print(f"Total Clauses: {finance_risk_summary['total_clauses_identified']}")
print(f"Financial Impact: {finance_risk_summary['financial_impact']}")

print(f"\nFinance Areas Covered:")
if finance_risk_summary['finance_areas']:
    for area in finance_risk_summary['finance_areas']:
        print(f"   - {area}")
else:
    print("   - No specific areas identified")

print(f"\nRisk Factors:")
for factor in finance_risk_summary['risk_factors']:
    print(f"   - {factor}")

print(f"\nRecommendations:")
for rec in finance_risk_summary['recommendations']:
    print(f"   - {rec}")

print("\n" + "="*70)
print("STEP 6 COMPLETE")
print("="*70)
```

```python
# LEGAL PIPELINE - STEP 6: Legal Risk Summary

print("\n" + "="*70)
print("LEGAL PIPELINE - STEP 6: LEGAL RISK SUMMARY")
print("="*70)

def generate_legal_risk_summary(validated_output):
    """Generate comprehensive legal risk summary."""
    summary = {
        "overall_risk": validated_output['risk_level'],
        "confidence_score": validated_output['confidence'],
        "total_clauses_identified": len(validated_output['extracted_clauses']),
        "legal_areas": [],
        "risk_factors": [],
        "recommendations": [],
        "legal_impact": "Unknown"
    }
    
    # Identify legal areas
    legal_areas = set()
    termination_count = 0
    liability_count = 0
    indemnification_count = 0
    
    for clause in validated_output['extracted_clauses']:
        clause_lower = clause.lower()
        
        if 'termination' in clause_lower:
            legal_areas.add("Termination")
            termination_count += 1
        if 'breach' in clause_lower:
            legal_areas.add("Breach of Contract")
        if 'governing law' in clause_lower or 'jurisdiction' in clause_lower:
            legal_areas.add("Governing Law & Jurisdiction")
        if 'liability' in clause_lower:
            legal_areas.add("Liability")
            liability_count += 1
        if 'indemnif' in clause_lower:
            legal_areas.add("Indemnification")
            indemnification_count += 1
        if 'intellectual property' in clause_lower:
            legal_areas.add("Intellectual Property")
        if 'confidential' in clause_lower:
            legal_areas.add("Confidentiality")
        if 'force majeure' in clause_lower:
            legal_areas.add("Force Majeure")
        if 'dispute' in clause_lower or 'arbitration' in clause_lower:
            legal_areas.add("Dispute Resolution")
    
    summary['legal_areas'] = list(legal_areas)
    
    # Risk factors and legal impact
    if validated_output['risk_level'] == 'high':
        summary['risk_factors'].append(f"High legal risk identified")
        summary['recommendations'].append("Immediate legal counsel review required")
        summary['recommendations'].append("Negotiate more favorable terms")
        summary['legal_impact'] = "HIGH - Significant legal exposure"
        
        if termination_count >= 2:
            summary['risk_factors'].append(f"Multiple termination clauses ({termination_count})")
            summary['recommendations'].append("Review termination rights carefully")
        
        if liability_count > 0 and indemnification_count == 0:
            summary['risk_factors'].append("Liability clauses without indemnification protection")
            summary['recommendations'].append("Request mutual indemnification clause")
    
    elif validated_output['risk_level'] == 'medium':
        summary['risk_factors'].append("Moderate legal obligations")
        summary['recommendations'].append("Review key clauses with legal team")
        summary['recommendations'].append("Ensure compliance processes are in place")
        summary['legal_impact'] = "MEDIUM - Standard legal terms with some exposure"
    
    else:
        summary['risk_factors'].append("Standard legal terms")
        summary['recommendations'].append("Maintain contractual compliance")
        summary['legal_impact'] = "LOW - Favorable legal conditions"
    
    if validated_output['confidence'] < 0.7:
        summary['risk_factors'].append(f"Low confidence score ({validated_output['confidence']:.0%})")
        summary['recommendations'].append("Manual legal review recommended")
    
    return summary

# Generate legal risk summary
legal_risk_summary = generate_legal_risk_summary(validated_legal)

print("\nLEGAL RISK SUMMARY")
print("="*70)

print(f"\nOverall Risk: {legal_risk_summary['overall_risk'].upper()}")
print(f"Confidence Score: {legal_risk_summary['confidence_score']:.0%}")
print(f"Total Clauses: {legal_risk_summary['total_clauses_identified']}")
print(f"Legal Impact: {legal_risk_summary['legal_impact']}")

print(f"\nLegal Areas Covered:")
if legal_risk_summary['legal_areas']:
    for area in legal_risk_summary['legal_areas']:
        print(f"   - {area}")
else:
    print("   - No specific areas identified")

print(f"\nRisk Factors:")
for factor in legal_risk_summary['risk_factors']:
    print(f"   - {factor}")

print(f"\nRecommendations:")
for rec in legal_risk_summary['recommendations']:
    print(f"   - {rec}")

print("\n" + "="*70)
print("STEP 6 COMPLETE")
print("="*70)
```

```python
# OPERATIONS PIPELINE - STEP 6: Operations Risk Summary

print("\n" + "="*70)
print("OPERATIONS PIPELINE - STEP 6: OPERATIONS RISK SUMMARY")
print("="*70)

def generate_operations_risk_summary(validated_output):
    """Generate comprehensive operations risk summary."""
    summary = {
        "overall_risk": validated_output['risk_level'],
        "confidence_score": validated_output['confidence'],
        "total_clauses_identified": len(validated_output['extracted_clauses']),
        "operations_areas": [],
        "risk_factors": [],
        "recommendations": [],
        "operational_impact": "Unknown"
    }
    
    # Identify operations areas
    operations_areas = set()
    sla_count = 0
    milestone_count = 0
    uptime_count = 0
    
    for clause in validated_output['extracted_clauses']:
        clause_lower = clause.lower()
        
        if 'sla' in clause_lower or 'service level' in clause_lower:
            operations_areas.add("Service Level Agreement")
            sla_count += 1
        if 'uptime' in clause_lower or 'availability' in clause_lower:
            operations_areas.add("Uptime & Availability")
            uptime_count += 1
        if 'response time' in clause_lower:
            operations_areas.add("Response Time")
        if 'milestone' in clause_lower:
            operations_areas.add("Project Milestones")
            milestone_count += 1
        if 'deliverable' in clause_lower:
            operations_areas.add("Deliverables")
        if 'performance' in clause_lower or 'metric' in clause_lower:
            operations_areas.add("Performance Metrics")
        if 'maintenance' in clause_lower:
            operations_areas.add("Maintenance")
        if 'timeline' in clause_lower or 'schedule' in clause_lower:
            operations_areas.add("Timeline & Schedule")
    
    summary['operations_areas'] = list(operations_areas)
    
    # Risk factors and operational impact
    if validated_output['risk_level'] == 'high':
        summary['risk_factors'].append("High execution risk identified")
        summary['recommendations'].append("Immediate operational review required")
        summary['recommendations'].append("Negotiate more achievable SLAs")
        summary['operational_impact'] = "HIGH - Significant execution challenges"
        
        if sla_count == 0:
            summary['risk_factors'].append("No clear SLA commitments")
            summary['recommendations'].append("Request explicit SLA definitions")
    
    elif validated_output['risk_level'] == 'medium':
        summary['risk_factors'].append("Moderate operational obligations")
        summary['recommendations'].append("Monitor SLA compliance closely")
        summary['recommendations'].append("Establish clear milestone tracking")
        summary['operational_impact'] = "MEDIUM - Standard operational requirements"
        
        if milestone_count >= 3:
            summary['risk_factors'].append(f"Multiple milestones ({milestone_count}) require tracking")
    
    else:
        summary['risk_factors'].append("Well-defined operational terms")
        summary['recommendations'].append("Maintain operational excellence")
        summary['operational_impact'] = "LOW - Achievable operational goals"
        
        if uptime_count > 0:
            summary['risk_factors'].append("Strong uptime commitments in place")
    
    if validated_output['confidence'] < 0.7:
        summary['risk_factors'].append(f"Low confidence score ({validated_output['confidence']:.0%})")
        summary['recommendations'].append("Manual operations review recommended")
    
    return summary

# Generate operations risk summary
operations_risk_summary = generate_operations_risk_summary(validated_operations)

print("\nOPERATIONS RISK SUMMARY")
print("="*70)

print(f"\nOverall Risk: {operations_risk_summary['overall_risk'].upper()}")
print(f"Confidence Score: {operations_risk_summary['confidence_score']:.0%}")
print(f"Total Clauses: {operations_risk_summary['total_clauses_identified']}")
print(f"Operational Impact: {operations_risk_summary['operational_impact']}")

print(f"\nOperations Areas Covered:")
if operations_risk_summary['operations_areas']:
    for area in operations_risk_summary['operations_areas']:
        print(f"   - {area}")
else:
    print("   - No specific areas identified")

print(f"\nRisk Factors:")
for factor in operations_risk_summary['risk_factors']:
    print(f"   - {factor}")

print(f"\nRecommendations:")
for rec in operations_risk_summary['recommendations']:
    print(f"   - {rec}")

print("\n" + "="*70)
print("STEP 6 COMPLETE")
print("="*70)
```

```python
# COORDINATOR MERGE - STEP 4: Compute Overall Risk

print("\n" + "="*70)
print("COORDINATOR MERGE - STEP 4: COMPUTE OVERALL RISK")
print("="*70)

def compute_overall_risk(merged_output):
    """
    Compute overall risk level from all agents.
    """
    print("\nComputing overall risk assessment...")
    
    # Collect risk levels
    risk_levels = {
        "legal": merged_output["legal"]["risk_level"],
        "compliance": merged_output["compliance"]["risk_level"],
        "finance": merged_output["finance"]["risk_level"],
        "operations": merged_output["operations"]["risk_level"]
    }
    
    # Count risk distribution
    risk_counts = {"high": 0, "medium": 0, "low": 0, "unknown": 0}
    for risk in risk_levels.values():
        risk_counts[risk] = risk_counts.get(risk, 0) + 1
    
    # Determine overall risk
    if risk_counts["high"] >= 2:
        overall_risk = "high"
    elif risk_counts["high"] >= 1:
        overall_risk = "medium-high"
    elif risk_counts["medium"] >= 2:
        overall_risk = "medium"
    elif risk_counts["medium"] >= 1:
        overall_risk = "low-medium"
    else:
        overall_risk = "low"
    
    # Calculate total clauses
    total_clauses = sum(
        len(merged_output[agent]["extracted_clauses"]) 
        for agent in merged_output
    )
    
    # Calculate average confidence
    confidences = [
        merged_output[agent]["confidence"] 
        for agent in merged_output
    ]
    average_confidence = sum(confidences) / len(confidences)
    
    risk_assessment = {
        "overall_risk_level": overall_risk,
        "risk_distribution": risk_counts,
        "individual_risks": risk_levels,
        "total_clauses": total_clauses,
        "average_confidence": average_confidence
    }
    
    print(f"\n   Overall Risk: {overall_risk.upper()}")
    print(f"   Risk Distribution:")
    for risk, count in risk_counts.items():
        if count > 0:
            print(f"      {risk.upper()}: {count} agent(s)")
    print(f"   Total Clauses: {total_clauses}")
    print(f"   Average Confidence: {average_confidence:.0%}")
    
    return risk_assessment

# Compute overall risk
overall_risk_assessment = compute_overall_risk(merged_output)

print("\n" + "="*70)
print("STEP 4 COMPLETE")
print("="*70)
```

```python
# COORDINATOR MERGE - STEP 5: Generate Recommendations

print("\n" + "="*70)
print("COORDINATOR MERGE - STEP 5: GENERATE RECOMMENDATIONS")
print("="*70)

def generate_consolidated_recommendations(merged_output, risk_assessment):
    """
    Generate consolidated recommendations from all agents.
    """
    print("\nGenerating consolidated recommendations...")
    
    recommendations = []
    
    # Overall risk-based recommendations
    if risk_assessment["overall_risk_level"] in ["high", "medium-high"]:
        recommendations.append({
            "priority": "HIGH",
            "category": "Overall",
            "recommendation": "Comprehensive legal and compliance review required before contract execution"
        })
    
    # Agent-specific recommendations
    for agent, data in merged_output.items():
        risk = data["risk_level"]
        clause_count = len(data["extracted_clauses"])
        
        if risk == "high":
            recommendations.append({
                "priority": "HIGH",
                "category": agent.capitalize(),
                "recommendation": f"{agent.capitalize()} analysis identified high risk - immediate review of {clause_count} clauses required"
            })
        elif risk == "medium" and clause_count >= 5:
            recommendations.append({
                "priority": "MEDIUM",
                "category": agent.capitalize(),
                "recommendation": f"Review {clause_count} {agent} clauses for potential risk mitigation"
            })
    
    # Confidence-based recommendations
    low_confidence_agents = [
        agent for agent, data in merged_output.items() 
        if data["confidence"] < 0.7
    ]
    if low_confidence_agents:
        recommendations.append({
            "priority": "MEDIUM",
            "category": "Quality",
            "recommendation": f"Manual verification recommended for: {', '.join(low_confidence_agents)}"
        })
    
    # Comprehensive review recommendation
    if risk_assessment["total_clauses"] > 20:
        recommendations.append({
            "priority": "LOW",
            "category": "Documentation",
            "recommendation": f"Extensive contract ({risk_assessment['total_clauses']} clauses) - consider detailed clause-by-clause review"
        })
    
    print(f"\n   Total Recommendations: {len(recommendations)}")
    print(f"   HIGH Priority: {sum(1 for r in recommendations if r['priority'] == 'HIGH')}")
    print(f"   MEDIUM Priority: {sum(1 for r in recommendations if r['priority'] == 'MEDIUM')}")
    print(f"   LOW Priority: {sum(1 for r in recommendations if r['priority'] == 'LOW')}")
    
    return recommendations

# Generate recommendations
consolidated_recommendations = generate_consolidated_recommendations(
    merged_output,
    overall_risk_assessment
)

print("\nRecommendations Generated:")
for i, rec in enumerate(consolidated_recommendations, 1):
    print(f"\n{i}. [{rec['priority']}] {rec['category']}")
    print(f"   {rec['recommendation']}")

print("\n" + "="*70)
print("STEP 5 COMPLETE")
print("="*70)
```

```python
# COORDINATOR MERGE - STEP 6: Build Final JSON Output

print("\n" + "="*70)
print("COORDINATOR MERGE - STEP 6: BUILD FINAL JSON OUTPUT")
print("="*70)

from datetime import datetime

# Build complete final output
final_merged_output = {
    "contract_analysis": {
        "legal": merged_output["legal"],
        "compliance": merged_output["compliance"],
        "finance": merged_output["finance"],
        "operations": merged_output["operations"]
    },
    "overall_risk_assessment": overall_risk_assessment,
    "recommendations": consolidated_recommendations,
    "metadata": {
        "analysis_date": datetime.now().isoformat(),
        "pipelines_executed": ["legal", "compliance", "finance", "operations"],
        "total_execution_time": sum([
            legal_output["metadata"].get("execution_time", 0),
            compliance_output["metadata"].get("execution_time", 0),
            finance_output["metadata"].get("execution_time", 0),
            operations_output["metadata"].get("execution_time", 0)
        ]) if all("execution_time" in x["metadata"] for x in [legal_output, compliance_output, finance_output, operations_output]) else 0.0
    }
}

print("\nFinal JSON Output Structure:")
print(f"   contract_analysis: 4 agents")
print(f"   overall_risk_assessment: {len(overall_risk_assessment)} metrics")
print(f"   recommendations: {len(consolidated_recommendations)} items")
print(f"   metadata: 3 fields")

print("\nOutput Summary:")
print(f"   Overall Risk: {final_merged_output['overall_risk_assessment']['overall_risk_level'].upper()}")
print(f"   Total Clauses: {final_merged_output['overall_risk_assessment']['total_clauses']}")
print(f"   Average Confidence: {final_merged_output['overall_risk_assessment']['average_confidence']:.0%}")
print(f"   Recommendations: {len(final_merged_output['recommendations'])}")

print("\n" + "="*70)
print("STEP 6 COMPLETE")
print("="*70)
```

```python
# COORDINATOR MERGE - STEP 7: Save Final Output

print("\n" + "="*70)
print("COORDINATOR MERGE - STEP 7: SAVE FINAL OUTPUT")
print("="*70)

import json
import os
from datetime import datetime

# Create Data directory
os.makedirs("Data", exist_ok=True)

# Save final merged output
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
final_output_file = f"Data/final_merged_output_{timestamp}.json"

with open(final_output_file, 'w') as f:
    json.dump(final_merged_output, f, indent=2)

print(f"\nFinal merged output saved to: {final_output_file}")

# Create summary report
summary_report = f"""
CONTRACT ANALYSIS SUMMARY REPORT
{'='*70}

Analysis Date: {final_merged_output['metadata']['analysis_date']}

OVERALL RISK ASSESSMENT
{'='*70}
Overall Risk Level: {final_merged_output['overall_risk_assessment']['overall_risk_level'].upper()}
Total Clauses Analyzed: {final_merged_output['overall_risk_assessment']['total_clauses']}
Average Confidence: {final_merged_output['overall_risk_assessment']['average_confidence']:.0%}

INDIVIDUAL AGENT RISKS
{'='*70}
"""

for agent, risk in final_merged_output['overall_risk_assessment']['individual_risks'].items():
    clauses = len(final_merged_output['contract_analysis'][agent]['extracted_clauses'])
    confidence = final_merged_output['contract_analysis'][agent]['confidence']
    summary_report += f"{agent.upper()}: {risk.upper()} ({clauses} clauses, {confidence:.0%} confidence)\n"

summary_report += f"""
RECOMMENDATIONS
{'='*70}
"""

for i, rec in enumerate(final_merged_output['recommendations'], 1):
    summary_report += f"\n{i}. [{rec['priority']}] {rec['category']}\n   {rec['recommendation']}\n"

# Save summary report
summary_file = f"Data/analysis_summary_{timestamp}.txt"
with open(summary_file, 'w') as f:
    f.write(summary_report)

print(f"Summary report saved to: {summary_file}")

print("\n" + summary_report)

print("\n" + "="*70)
print("STEP 7 COMPLETE")
print("="*70)
```

```python
# COORDINATOR MERGE - STUDENT TASK

print("\n\n" + "="*70)
print("STUDENT TASK: CONFIDENCE AGGREGATION & HIGHEST-RISK CLAUSE")
print("="*70)

print("\nTask:")
print("   1. Add confidence aggregation logic")
print("   2. Print highest-risk clause")

# TASK 1: Confidence Aggregation Logic
print("\n" + "="*70)
print("TASK 1: CONFIDENCE AGGREGATION LOGIC")
print("="*70)

def aggregate_confidence_metrics(merged_output):
    """
    Aggregate detailed confidence metrics from all agents.
    """
    print("\nAggregating confidence metrics...")
    
    confidence_metrics = {
        "by_agent": {},
        "overall_average": 0.0,
        "weighted_average": 0.0,
        "highest_confidence": {"agent": "", "score": 0.0},
        "lowest_confidence": {"agent": "", "score": 1.0},
        "confidence_distribution": {"high": 0, "medium": 0, "low": 0}
    }
    
    # Collect agent confidences
    total_clauses = 0
    weighted_sum = 0
    
    for agent, data in merged_output.items():
        confidence = data["confidence"]
        clause_count = len(data["extracted_clauses"])
        
        confidence_metrics["by_agent"][agent] = {
            "confidence": confidence,
            "clause_count": clause_count
        }
        
        # Track highest and lowest
        if confidence > confidence_metrics["highest_confidence"]["score"]:
            confidence_metrics["highest_confidence"] = {"agent": agent, "score": confidence}
        if confidence < confidence_metrics["lowest_confidence"]["score"]:
            confidence_metrics["lowest_confidence"] = {"agent": agent, "score": confidence}
        
        # Calculate weighted average
        weighted_sum += confidence * clause_count
        total_clauses += clause_count
        
        # Confidence distribution
        if confidence >= 0.8:
            confidence_metrics["confidence_distribution"]["high"] += 1
        elif confidence >= 0.6:
            confidence_metrics["confidence_distribution"]["medium"] += 1
        else:
            confidence_metrics["confidence_distribution"]["low"] += 1
    
    # Calculate averages
    confidences = [data["confidence"] for data in merged_output.values()]
    confidence_metrics["overall_average"] = sum(confidences) / len(confidences)
    confidence_metrics["weighted_average"] = weighted_sum / total_clauses if total_clauses > 0 else 0
    
    return confidence_metrics

# Aggregate confidence
confidence_aggregation = aggregate_confidence_metrics(merged_output)

print("\nCONFIDENCE AGGREGATION RESULTS")
print("="*70)

print(f"\nOverall Average Confidence: {confidence_aggregation['overall_average']:.0%}")
print(f"Weighted Average Confidence: {confidence_aggregation['weighted_average']:.0%}")

print(f"\nHighest Confidence:")
print(f"   Agent: {confidence_aggregation['highest_confidence']['agent'].upper()}")
print(f"   Score: {confidence_aggregation['highest_confidence']['score']:.0%}")

print(f"\nLowest Confidence:")
print(f"   Agent: {confidence_aggregation['lowest_confidence']['agent'].upper()}")
print(f"   Score: {confidence_aggregation['lowest_confidence']['score']:.0%}")

print(f"\nConfidence by Agent:")
for agent, metrics in confidence_aggregation['by_agent'].items():
    print(f"   {agent.upper()}: {metrics['confidence']:.0%} ({metrics['clause_count']} clauses)")

print(f"\nConfidence Distribution:")
for level, count in confidence_aggregation['confidence_distribution'].items():
    print(f"   {level.upper()}: {count} agent(s)")

# TASK 2: Print Highest-Risk Clause
print("\n" + "="*70)
print("TASK 2: IDENTIFY HIGHEST-RISK CLAUSE")
print("="*70)

def find_highest_risk_clause(merged_output):
    """
    Find the highest-risk clause across all agents.
    """
    print("\nIdentifying highest-risk clause...")
    
    # Risk level priority
    risk_priority = {"high": 3, "medium": 2, "low": 1, "unknown": 0}
    
    highest_risk_clause = {
        "agent": "",
        "risk_level": "unknown",
        "clause": "",
        "confidence": 0.0
    }
    
    for agent, data in merged_output.items():
        agent_risk = data["risk_level"]
        agent_confidence = data["confidence"]
        clauses = data["extracted_clauses"]
        
        if clauses and risk_priority[agent_risk] > risk_priority[highest_risk_clause["risk_level"]]:
            highest_risk_clause = {
                "agent": agent,
                "risk_level": agent_risk,
                "clause": clauses[0],  # First clause as representative
                "confidence": agent_confidence
            }
    
    return highest_risk_clause

# Find highest-risk clause
highest_risk = find_highest_risk_clause(merged_output)

print("\nHIGHEST-RISK CLAUSE IDENTIFIED")
print("="*70)

print(f"\nAgent: {highest_risk['agent'].upper()}")
print(f"Risk Level: {highest_risk['risk_level'].upper()}")
print(f"Confidence: {highest_risk['confidence']:.0%}")

print(f"\nClause:")
print("-" * 70)
print(f"{highest_risk['clause']}")
print("-" * 70)

print(f"\nRisk Context:")
print(f"   This {highest_risk['agent']} clause has {highest_risk['risk_level']} risk")
print(f"   Analysis confidence: {highest_risk['confidence']:.0%}")
print(f"   Recommendation: Immediate review by {highest_risk['agent']} team")

# Add to final output
final_merged_output["confidence_aggregation"] = confidence_aggregation
final_merged_output["highest_risk_clause"] = highest_risk

# Save updated output
updated_output_file = f"Data/final_merged_output_with_aggregation_{timestamp}.json"
with open(updated_output_file, 'w') as f:
    json.dump(final_merged_output, f, indent=2)

print(f"\nUpdated output saved to: {updated_output_file}")

print("\n" + "="*70)
print("STUDENT TASK COMPLETE")
print("="*70)

print("\nTask Summary:")
print("   [X] Confidence aggregation logic added")
print("   [X] Overall average confidence calculated")
print("   [X] Weighted average confidence calculated")
print("   [X] Highest confidence agent identified")
print("   [X] Lowest confidence agent identified")
print("   [X] Confidence distribution computed")
print("   [X] Highest-risk clause identified")
print("   [X] Risk context provided")
print("   [X] Updated output saved")

print("\n" + "="*70)
print("COORDINATOR MERGE - ALL TASKS COMPLETE!")
print("="*70)

print("\n🎉 FINAL MODULE COMPLETE! 🎉")
print("\nYou have successfully completed:")
print("   ✓ Coordinator Merging Module")
print("   ✓ All pipeline outputs merged")
print("   ✓ Overall risk assessment")
print("   ✓ Consolidated recommendations")
print("   ✓ Confidence aggregation")
print("   ✓ Highest-risk clause identification")

print("\n" + "="*70)
```
